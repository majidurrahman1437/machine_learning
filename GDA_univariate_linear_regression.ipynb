```python
print("This algorithm is intended for executing Gradient Descent Algorithm for Univariate Linear Regression.\n")
print("Here, we have assumed x as independent variable and y as dependent variable.\n")

num_of_data = int(input("Enter the number of data points:"))
x = list()
y = list()
hw_x = list()
error = list()
squared_error = list()
w0 = 0.0
w1 = 0.0
alpha = 0.01
count = 0
previous_cost = 100000.00
cost_difference = 100000.00
for i in range(num_of_data):
    print("X[",i,"] = ")
    x.append(float(input()))
    print("Y[",i,"] = ")
    y.append(float(input()))
    
print("Feature Dataset = ", x)
print("Original Output = ", y)
print("\n\n")

while(cost_difference >=  0.001):
    for a in range(num_of_data):
        sum_of_error = 0.0
        sum_of_squared_error = 0.0
        for j in range(0, num_of_data):
            if (a == 0 and count == 0):
                hw_x.append(w0 + (w1 * x[j]))
                error.append(hw_x[j] - y[j])
                squared_error.append(error[j] * error[j]) 
            else:
                hw_x[j] = w0 + (w1 * x[j])
                error[j] = hw_x[j] - y[j]
                squared_error[j] = error[j] * error[j]
            
            sum_of_error = sum_of_error + error[j]
            sum_of_squared_error = sum_of_squared_error + squared_error[j]

        cost = sum_of_squared_error / (2 * num_of_data)
        cost_difference = previous_cost - cost
        previous_cost = cost
        w0 = w0 - (alpha * (sum_of_error / num_of_data))
        w1 = w1 - (alpha * (sum_of_error / num_of_data) * x[a])
        print("Estimated Cost for Round-", count+1," and data point-", a+1," = ", cost)
        print("Updated weight_0 for Round-", count+1," and data point-", a+1," = ", w0)
        print("Updated weight_1 for Round-", count+1," and data point-", a+1," = ", w1)
        print("Predicted output for Round-", count+1," and data point-", a+1," = ", hw_x)
        print("Error for Round-", count+1," and data point-", a+1," = ", error)
        print("Sum of errors for Round-", count+1," and data point-", a+1," = ", sum_of_error)
        print("Squared Error for Round-", count+1," and data point-", a+1," = ", squared_error)
        print("Sum of squared errors for Round-", count+1," and data point-", a+1," = ", sum_of_squared_error)
        print("\n")    
    count = count + 1
    
    
print("\n\n")
print("The final hypothesis is: hw_x = ", w0, " + ", w1, " * x")
```
