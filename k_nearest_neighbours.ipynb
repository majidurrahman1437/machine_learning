```python
import math
from operator import itemgetter

print("This algorithm is intended for executing k-nearest neighbours algorithm.\n")

num_of_data = int(input("Enter the number of datapoints: "))
num_of_features = int(input("Enter the number of features: "))
unknown_point = list()
print("Enter the unknown datapoint:")
for x in range(num_of_features):
    unknown_point.append(float(input()))
    
value_of_k = int(math.sqrt(num_of_data))
if((value_of_k % 2) == 0):
    value_of_k = value_of_k + 1
    
dataset = []
for i in range(num_of_data):
    sq_dist = 0.0
    print("Dataset -", i)
    dataset.append([])
    for j in range(num_of_features + 2):
        if(j < num_of_features):
            dataset[i].append(float(input()))
            sq_dist = sq_dist + ((dataset[i][j] - unknown_point[j]) * (dataset[i][j] - unknown_point[j]))
        elif(j == num_of_features):
            dataset[i].append(float(input()))
        else:
            dataset[i].append(math.sqrt(sq_dist))
        
print("\nDataset = ", dataset)
print("\nThe value of k:", value_of_k)
dataset = sorted(dataset, key=itemgetter(num_of_features + 1))
print("\nDataset after distance-wise sorting = ", dataset)

count0 = 0
count1 = 0
for k in range(value_of_k):
    if(dataset[k][num_of_features] == 0.0):
        count0 = count0 + 1
    else:
        count1 = count1 + 1
        
if(count0> count1):
    print("\nThe datapoint belongs to class-0(CAT)")
else:
    print("\nThe datapoint belongs to class-1(DOG)")
```
